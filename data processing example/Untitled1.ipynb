```python
# -*- coding: utf-8 -*
import matplotlib.pyplot as mp
import pandas as pd
import numpy as np
from bokeh.plotting import figure, output_file, show


# 讀檔(來源)
def read(location, locationS):
    src = f'Record/106年 {location}空品區/106年{locationS}站_20180309.xls'
    df = pd.read_excel(src)
    return df

def data_type(df, name):
    def add_to_list(series):
        list1 = []
        for n in range(0, 24):
            if type(series[n+3]) != str:
                list1.append(series[n+3])
            else:
                list1.append(float('nan'))
        return list1

    rows = df.shape[0]
    list1 = []
    for n in range(0, rows):
        if df.loc[n][2] == name:
            list1.append(add_to_list(df.loc[n]))
    return list1

def average(list_ori):
    sum = 0
    amount = 0
    for item in list_ori:
        if np.isnan(item):
            sum += 0
        else:
            sum += item
            amount += 1
    return sum/amount

def draw(location, list1, height):
    listX = np.arange(1, len(list1)+1)
  
    mp.ylim(0, height)
    mp.plot(listX, list1, label = location)
    mp.legend(loc = 'upper right')
    mp.show()

def rd(location, locationS):
    df = read(location, locationS)
    data_pm25 = data_type(df, 'PM2.5')

    list1 = [[]for y in range(24)]
    for n in range(0, 24):
        for m in range(0, len(data_pm25)):
            list1[n].append(data_pm25[m][n])
    list_out = []
    print(list1)
    for li in list1:
        list_out.append(average(li))
    print(list_out)
    a = np.arange(24)
    mp.plot(a, list_out)
    mp.xlabel('hour')
    mp.ylabel('μg/m3')
    mp.ylim(0, 40)
    print(location, locationS)
    mp.show()

#     output_file("out.html")
#     p = figure()
#     p.line(a,list_out)
#     show(p)
rd('北部','三重')
```
