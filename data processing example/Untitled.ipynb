```python
%matplotlib inline
import matplotlib.pyplot as mp
import pandas as pd
import math

def readraw(src):
    def average(numList):
        if(len(numList) != 0):
            sum = 0.0
            for item in numList:
                sum += item
            return sum/len(numList)
        else:
            return 0
    df = pd.read_excel(src)
    rows, columns = df.shape
    count = 0
    ary = [[]for y in range(365)]
    for n in range(0, rows):
        if df.loc[n]['測項'] == 'PM2.5':
            for item in df.loc[n]:
                if type(item) != str and math.isnan(item) == False:
                    ary[count].append(float(item))
            count += 1
    list1 = []
    for x in range(0, len(ary)):
        for item in ary[x]:
            if math.isnan(item):
                ary[x].remove(item)
    for n in range(0, len(ary)):
        list1.append(average(ary[n]))
    listX = []
    for n in range(0, len(list1)):
        listX.append(n)
    mp.ylim(0, 100)
    mp.plot(listX, list1, label = 'one')
```
